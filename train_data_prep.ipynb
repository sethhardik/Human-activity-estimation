```python
import numpy as np
import pandas as pd
import glob
```

```python
# importing all the dataset

path_phone_accel = "train\\phone\\accel\\" 
df_phone_accel =pd.DataFrame()
all_files = glob.glob(path_phone_accel + "\\*.txt")
for file in all_files:
    df = pd.read_csv(file,header = None)
    df_phone_accel = pd.concat([df_phone_accel,df])

path_phone_gyro = "train\\phone\\gyro\\" 
df_phone_gyro = pd.DataFrame()
all_files = glob.glob(path_phone_gyro + "\\*.txt")
for file in all_files:
    df = pd.read_csv(file,header = None)
    df_phone_gyro = pd.concat([df_phone_gyro,df])

path_watch_accel = "train\\watch\\accel\\"
df_watch_accel =pd.DataFrame()
all_files = glob.glob(path_watch_accel + "\\*.txt")
for file in all_files:
    df = pd.read_csv(file,header = None)
    df_watch_accel = pd.concat([df_watch_accel,df])

path_watch_gyro = "train\\watch\\gyro\\" 
df_watch_gyro = pd.DataFrame()
all_files = glob.glob(path_watch_gyro + "\\*.txt")
for file in all_files:
    df = pd.read_csv(file,header = None)
    df_watch_gyro = pd.concat([df_watch_gyro,df])
```

```python
# giving names to the columns of all the dataset that we have with us 
df_phone_accel.columns =['ID','output','timestamp','accel_phone_x','accel_phone_y','accel_phone_z']
df_phone_gyro.columns =['ID','output','timestamp','gyro_phone_x','gyro_phone_y','gyro_phone_z']

df_watch_accel.columns =['ID','output','timestamp','accel_watch_x','accel_watch_y','accel_watch_z']
df_watch_gyro.columns =['ID','output','timestamp','gyro_watch_x','gyro_watch_y','gyro_watch_z']
```

```python
# removing semicolon from the csv files 
df_phone_accel["accel_phone_z"] = df_phone_accel["accel_phone_z"].str.replace(";","")
df_phone_accel["accel_phone_z"] = pd.to_numeric(df_phone_accel["accel_phone_z"])

df_phone_gyro["gyro_phone_z"] = df_phone_gyro["gyro_phone_z"].str.replace(";","")
df_phone_gyro["gyro_phone_z"] = pd.to_numeric(df_phone_gyro["gyro_phone_z"])

df_watch_accel["accel_watch_z"] = df_watch_accel["accel_watch_z"].str.replace(";","")
df_watch_accel["accel_watch_z"] = pd.to_numeric(df_watch_accel["accel_watch_z"])

df_watch_gyro["gyro_watch_z"] = df_watch_gyro["gyro_watch_z"].str.replace(";","")
df_watch_gyro["gyro_watch_z"] = pd.to_numeric(df_watch_gyro["gyro_watch_z"])
```

```python
# dropping timestamp as it won't help in predictions
df_phone_accel.drop(["timestamp"],axis = 1, inplace = True)
df_phone_gyro.drop(["timestamp"],axis = 1, inplace = True)
df_watch_accel.drop(["timestamp"],axis = 1, inplace = True)
df_watch_gyro.drop(["timestamp"],axis = 1, inplace = True)

df_phone_accel.head()
```

```python
df_watch_gyro.head()
```

```python
# saving mean and variance of values in seperate dataframe not using standard deviation in this step as it gave an unsolvable error
df_phone_accel_mean= df_phone_accel.groupby(["ID","output"],as_index =False).mean()
df_phone_accel_mean.columns = ["ID","output",'accel_phone_x_mean','accel_phone_y_mean','accel_phone_z_mean']
df_phone_gyro_mean= df_phone_gyro.groupby(["ID","output"],as_index =False).mean()
df_phone_gyro_mean.columns = ["ID","output",'gyro_phone_x_mean','gyro_phone_y_mean','gyro_phone_z_mean']
df_watch_accel_mean= df_watch_accel.groupby(["ID","output"],as_index =False).mean()
df_watch_accel_mean.columns = ["ID","output",'accel_watch_x_mean','accel_watch_y_mean','accel_watch_z_mean']
df_watch_gyro_mean= df_watch_gyro.groupby(["ID","output"],as_index =False).mean()
df_watch_gyro_mean.columns = ["ID","output",'gyro_watch_x_mean','gyro_watch_y_mean','gyro_watch_z_mean']
# variance

df_phone_accel_std= df_phone_accel.groupby(["ID","output"],as_index =False).var()
df_phone_accel_std.columns = ["ID","output",'accel_phone_x_std','accel_phone_y_std','accel_phone_z_std']
df_phone_gyro_std= df_phone_gyro.groupby(["ID","output"],as_index =False).var()
df_phone_gyro_std.columns = ["ID","output",'gyro_phone_x_std','gyro_phone_y_std','gyro_phone_z_std']
df_watch_accel_std= df_watch_accel.groupby(["ID","output"],as_index =False).var()
df_watch_accel_std.columns = ["ID","output",'accel_watch_x_std','accel_watch_y_std','accel_watch_z_std']
df_watch_gyro_std= df_watch_gyro.groupby(["ID","output"],as_index =False).var()
df_watch_gyro_std.columns = ["ID","output",'gyro_watch_x_std','gyro_watch_y_std','gyro_watch_z_std']

# max
df_phone_accel_max= df_phone_accel.groupby(["ID","output"],as_index =False).max()
df_phone_accel_max.columns = ["ID","output",'accel_phone_x_max','accel_phone_y_max','accel_phone_z_max']
df_phone_gyro_max= df_phone_gyro.groupby(["ID","output"],as_index =False).max()
df_phone_gyro_max.columns = ["ID","output",'gyro_phone_x_max','gyro_phone_y_max','gyro_phone_z_max']
df_watch_accel_max= df_watch_accel.groupby(["ID","output"],as_index =False).max()
df_watch_accel_max.columns = ["ID","output",'accel_watch_x_max','accel_watch_y_max','accel_watch_z_max']
df_watch_gyro_max= df_watch_gyro.groupby(["ID","output"],as_index =False).max()
df_watch_gyro_max.columns = ["ID","output",'gyro_watch_x_max','gyro_watch_y_max','gyro_watch_z_max']


# max
df_phone_accel_min= df_phone_accel.groupby(["ID","output"],as_index =False).min()
df_phone_accel_min.columns = ["ID","output",'accel_phone_x_min','accel_phone_y_min','accel_phone_z_min']
df_phone_gyro_min= df_phone_gyro.groupby(["ID","output"],as_index =False).min()
df_phone_gyro_min.columns = ["ID","output",'gyro_phone_x_min','gyro_phone_y_min','gyro_phone_z_min']
df_watch_accel_min= df_watch_accel.groupby(["ID","output"],as_index =False).min()
df_watch_accel_min.columns = ["ID","output",'accel_watch_x_min','accel_watch_y_min','accel_watch_z_min']
df_watch_gyro_min= df_watch_gyro.groupby(["ID","output"],as_index =False).min()
df_watch_gyro_min.columns = ["ID","output",'gyro_watch_x_min','gyro_watch_y_min','gyro_watch_z_min']
```

```python
# checikng shape of the datasets that we have with us 
print(df_watch_gyro_mean.shape)
print(df_watch_accel_mean.shape)
print(df_phone_gyro_mean.shape)
print(df_phone_accel_mean.shape)

print(df_watch_gyro_std.shape)
print(df_watch_accel_std.shape)
print(df_phone_gyro_std.shape)
print(df_phone_accel_std.shape)

print(df_watch_gyro_max.shape)
print(df_watch_accel_max.shape)
print(df_phone_gyro_max.shape)
print(df_phone_accel_max.shape)

print(df_watch_gyro_min.shape)
print(df_watch_accel_min.shape)
print(df_phone_gyro_min.shape)
print(df_phone_accel_min.shape)
```

```python
# creating seperate files of mean and variances in order to merge them later 
df_phone_mean = pd.concat([df_phone_accel_mean,df_phone_gyro_mean["gyro_phone_x_mean"],df_phone_gyro_mean["gyro_phone_y_mean"],df_phone_gyro_mean["gyro_phone_z_mean"]],axis=1)
df_phone_std = pd.concat([df_phone_accel_std,df_phone_gyro_std["gyro_phone_x_std"],df_phone_gyro_std["gyro_phone_y_std"],df_phone_gyro_std["gyro_phone_z_std"]],axis=1)
df_watch_mean = pd.concat([df_watch_accel_mean,df_watch_gyro_mean["gyro_watch_x_mean"],df_watch_gyro_mean["gyro_watch_y_mean"],df_watch_gyro_mean["gyro_watch_z_mean"]],axis=1)
df_watch_std = pd.concat([df_watch_accel_std,df_watch_gyro_std["gyro_watch_x_std"],df_watch_gyro_std["gyro_watch_y_std"],df_watch_gyro_std["gyro_watch_z_std"]],axis=1)

df_phone_max = pd.concat([df_phone_accel_max,df_phone_gyro_max["gyro_phone_x_max"],df_phone_gyro_max["gyro_phone_y_max"],df_phone_gyro_max["gyro_phone_z_max"]],axis=1)
df_watch_max = pd.concat([df_watch_accel_max,df_watch_gyro_max["gyro_watch_x_max"],df_watch_gyro_max["gyro_watch_y_max"],df_watch_gyro_max["gyro_watch_z_max"]],axis=1)

df_phone_min = pd.concat([df_phone_accel_min,df_phone_gyro_min["gyro_phone_x_min"],df_phone_gyro_min["gyro_phone_y_min"],df_phone_gyro_min["gyro_phone_z_min"]],axis=1)
df_watch_min = pd.concat([df_watch_accel_min,df_watch_gyro_min["gyro_watch_x_min"],df_watch_gyro_min["gyro_watch_y_min"],df_watch_gyro_min["gyro_watch_z_min"]],axis=1)
```

```python
df_watch_std.info()
```

```python
# creating single file having all the columns and droping duplicate of ID and Output
# this step made all the columns to object datatype which we later convert to numerical 
df_all = pd.concat([df_phone_mean,df_watch_mean,df_watch_std,df_phone_std,df_phone_max,df_watch_max,df_phone_min,df_watch_min],axis =1).T.drop_duplicates().T
```

```python
# droping rows containing NAN in them as there are less no need to fill them 
df_no_na = df_all.dropna()
```

```python
df_no_na.info()
```

```python
# saving id and output in different data frame 
df_temp = df_no_na.iloc[:,0:2]
df_temp
```

```python
# dropping id and output row due to some error there were 2 id and output and both gets removed thats why i merge the temp
# in next step
df_no_na.drop(df_no_na.iloc[:, 0:2], inplace = True, axis = 1)
df_no_na.head()
```

```python
# creating final dataset 
df = pd.concat([df_temp,df_no_na],axis=1)
```

```python
# converting all object files to numeric format 
df = df.apply(pd.to_numeric,errors='ignore') 
```

```python
df.info()
```

```python
# finally converting variance of rows to standard deviation 
df.loc[:,"accel_watch_x_std":"gyro_phone_z_std"] =np.sqrt(df.loc[:,"accel_watch_x_std":"gyro_phone_z_std"])
```
